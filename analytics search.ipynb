```python
import matplotlib.pyplot as plt
import pandas as pd
import json

file = "search.json"
data = None
with open(file, 'r') as f:
    data = json.load(f)


video_stats1 = data
#print(video_stats1)

```

```python
#video statistics


sorted_vids = sorted(video_stats1.items(), key=lambda item: int(item[1]["viewCount"]), reverse=True)
stats = []
for vid in sorted_vids:
    video_id = vid[0]
    title = vid[1]["title"]
    views = int(vid[1]["viewCount"])
    likes = int(vid[1].get("likeCount",0))
    comments = int(vid[1].get("commentCount",0))
    stats.append([title,views,likes,comments])
```

```python
#create pandas dataframe

df = pd.DataFrame(stats, columns=["title","views","likes","comments"])
df.head(10)
```

```python
top10 = df.head(10)
ax = top10.plot.bar(x="title", y="views", figsize=(12,8), fontsize=14)
```

```python
bottom10 = df.tail(10)
ax = bottom10.plot.bar(x="title", y="views", figsize=(12,8), fontsize=14)
```
